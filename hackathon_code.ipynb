### lets create a model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.multioutput import MultiOutputClassifier

from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_curve, roc_auc_score
RANDOM_SEED=6
```

```python
# now get only nueric columns
numeric_cols=df.columns[df.dtypes != "object"].values
print(numeric_cols)
```

```python
# chain preprocessing into a Pipeline object
numeric_preprocessing_steps = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('simple_imputer', SimpleImputer(strategy='median'))
])
# creating a precprocessor of final pipeline 
preprocessor = ColumnTransformer(
    transformers = [
        ("numeric", numeric_preprocessing_steps, numeric_cols)
    ],
    remainder = "drop"
)
```

```python
# for estimation 

estimators= MultiOutputClassifier(
estimator=LogisticRegression(penalty="l2",C=1))

## for full pipeline 
full_pipeline = Pipeline([
    ("preprocessor",preprocessor),("estimators",estimators)
])
```

```python
full_pipeline
```

```python
## now lets have some training 
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,labels_df,test_size=0.33,
                                               shuffle=True,
                                              stratify=labels_df,
                                               random_state=RANDOM_SEED)
```

```python
# Train model
full_pipeline.fit(X_train, y_train)

# Predict on evaluation set
preds = full_pipeline.predict_proba(X_test)
preds
```

```python
preds[0].shape
preds[1].shape
```

```python
y_preds = pd.DataFrame(
    {
        "xyz_vaccine": preds[0][:, 1],
        "seasonal_vaccine": preds[1][:, 1],
    },
    index = y_test.index
)
y_preds.head()
```

###for roc-auc curve 

```python
def plot_roc(y_true, y_score, label_name, ax):
    fpr, tpr, thresholds = roc_curve(y_true, y_score)
    ax.plot(fpr, tpr)
    ax.plot([0, 1], [0, 1], color='black', linestyle='--')
    ax.set_ylabel('True_positive')
    ax.set_xlabel('False_positive')
    ax.set_title(
        f"{label_name}: AUC = {roc_auc_score(y_true, y_score):.4f}"
    )
```

```python
fig, ax = plt.subplots(1, 2, figsize=(6, 3))

plot_roc(
    y_test['xyz_vaccine'], 
    y_preds['xyz_vaccine'], 
    'h1n1_vaccine',
    ax=ax[0]
)
plot_roc(
    y_test['seasonal_vaccine'], 
    y_preds['seasonal_vaccine'], 
    'seasonal_vaccine',
    ax=ax[1]
)
fig.tight_layout()
```
